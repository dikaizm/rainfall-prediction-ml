Kolom RainToday dan RainTomorrow bertipe data object (Yes/No), perlu diubah ke dalam bentuk binary (0 atau 1).

```python
df['RainToday'].replace({'No':0, 'Yes':1}, inplace=True)
df['RainTomorrow'].replace({'No':0, 'Yes':1}, inplace=True)
```

```python
df.head(10)
```

Cek apakah dataset balanced atau unbalanced. Jika unbalanced, perlu dilakukan downsample data mayoritas atau oversample minoritas untuk menyeimbangkannya.

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6,4))
df.RainTomorrow.value_counts(normalize=True).plot(kind='bar', color=['orange', 'green'], alpha=0.9, rot=0)
plt.title('RainTomorrow check imbalanced dataset (Yes=1)/(No=0)')
plt.show()
```
