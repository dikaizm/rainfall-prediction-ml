Kolom RainToday dan RainTomorrow bertipe data object (Yes/No), perlu diubah ke dalam bentuk binary (0 atau 1).

```python
df['RainToday'].replace({'No':0, 'Yes':1}, inplace=True)
df['RainTomorrow'].replace({'No':0, 'Yes':1}, inplace=True)
```

```python
df.head(10)
```

Cek apakah dataset balanced atau unbalanced. Jika unbalanced, perlu dilakukan downsample data mayoritas atau oversample minoritas untuk menyeimbangkannya.

```python
import matplotlib.pyplot as plt

norm = df.RainTomorrow.value_counts(normalize=True)
print(norm)

fig = plt.figure(figsize=(6,4))
norm.plot(kind='bar', color=['orange', 'green'], alpha=0.9, rot=0)
plt.title('RainTomorrow check imbalanced dataset (Yes=1)/(No=0)')
plt.show()
```

Terdapat dataset imbalanced dengan rasio 78:22. Karena ukuran dataset termasuk kecil, maka akan dilakukan oversampling dengan kelas minoritas.

```python
# Handle class imbalance
from sklearn.utils import resample

no = df[df.RainTomorrow == 0]
yes = df[df.RainTomorrow == 1]
yes_oversampled = resample(yes, replace=True, n_samples=len(no), random_state=72)
oversampled = pd.concat([no, yes_oversampled])

fig = plt.figure(figsize=(6,4))
norm = oversampled.RainTomorrow.value_counts(normalize=True)
print(norm)

norm.plot(kind='bar', color=['orange','green'], alpha=0.9, rot=0)
plt.title("RainTomorrow after oversampling")
plt.show()
```

```python
# Check missing data using heatmap
import seaborn as sns
sns.heatmap(oversampled.isnull(), cbar=False, cmap='PuBu')
```

```python
sum_null = oversampled.isnull().sum()
count_null = oversampled.isnull().count()

total = sum_null.sort_values(ascending=False)
percent = sum_null/count_null.sort_values(ascending=False)
missing = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing.head(4)
```

Hasil penemuan memperlihatkan dari 4 features memiliki missing value di bawah 50%, angka ini masih considerable sehingga data dapat dipakai dengan catatan perlu dilakukan proses imputation.

```python
# Imputation and transformation

# Find column with object data types
oversampled.select_dtypes(include=['object']).columns
```

```python
# Impute categorical variable with mode
oversampled['Date'].fillna(oversampled['Date'].mode()[0], inplace=True)
oversampled['Location'].fillna(oversampled['Location'].mode()[0], inplace=True)
oversampled['WindGustDir'].fillna(oversampled['WindGustDir'].mode()[0], inplace=True)
oversampled['WindDir9am'].fillna(oversampled['WindDir9am'].mode()[0], inplace=True)
oversampled['WindDir3pm'].fillna(oversampled['WindDir3pm'].mode()[0], inplace=True)
```

```python
# Convert categorical features to continuous features with LabelEncoding
from sklearn.preprocessing import LabelEncoder
label_encoder = {}
for col in oversampled.select_dtypes(include=['object']).columns:
    label_encoder[col] = LabelEncoder()
    oversampled[col] = label_encoder[col].fit_transform(oversampled[col])
```

```python
# Impute missing values with MICE (Multiple Imputation by Chained Equations)
import warnings
warnings.filterwarnings("ignore")

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
MiceImputed = oversampled.copy(deep=True)
mice_imputer = IterativeImputer()
MiceImputed.iloc[:,:] = mice_imputer.fit_transform(oversampled)
```

```python
# Check missing values again
MiceImputed.isnull().sum()
```

```python
# Detect and eliminate outliers using interquartile
Q1 = MiceImputed.quantile(0.25)
Q3 = MiceImputed.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
cols_no_date = MiceImputed.drop(columns=['Date']).columns

data_to_plot = [MiceImputed[col].dropna().values for col in cols_no_date]

# Boxplot IQR to visualize outliers
plt.figure(figsize=(12, 5))
plt.boxplot(x=data_to_plot, labels=cols_no_date)
plt.title("Boxplot columns")
plt.show()
```
